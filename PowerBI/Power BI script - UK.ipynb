```python
#import the tools needed
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# load the dataset: 
dataset = pd.read_excel('Sales.xlsx')
dataset.head()
```

```python
plt.scatter(dataset.TV, dataset.Sales)
plt.show()
```

```python
#define de components (X) and the predictions (y)
X = dataset.iloc[:, 1:4]
y = dataset['Sales']
X_train, X_test, y_train, y_test = train_test_split(X, y)

```

```python
# Construct a model and make the prediction on the testset
lin = LinearRegression()
lin.fit(X_train, y_train)
y_pred = lin.predict(X_test)

```

```python
# Make the coefficients (slope values)
coef = lin.coef_
components = pd.DataFrame(zip(X.columns, coef), columns= ['component', 'value'])
components = components.append({'component':'intercept', 'value': lin.intercept_}, ignore_index = True)
```

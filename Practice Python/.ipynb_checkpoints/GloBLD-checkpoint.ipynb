# Polygoon script voor GloBLD door BAIME

```python
# Voorbeeld met verzonnen data

import matplotlib.pyplot as plt
from matplotlib.collections import PolyCollection
import numpy as np

# some sample data
data = [(0,.5), (1,.7), (2,.3), (3,.6), (4,.2)]

fig = plt.figure()
ax = fig.add_subplot(111)

# draw the polygon
p = PolyCollection([data], facecolor=(1,0,0,.5), edgecolor='none')
ax.add_collection(p)

# draw the line
d = np.array(data)
ax.plot(*zip(*data), color='k', linewidth=2)
```

```python
# Voorbeeld 3D polygoon
from mpl_toolkits import mplot3d
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
from itertools import product, combinations
fig = plt.figure()
ax = plt.axes(projection='3d')

r = [-1, 1]
for s, e in combinations(np.array(list(product(r, r, r))), 2):
    if np.sum(np.abs(s-e)) == r[1]-r[0]:
        ax.plot3D(*zip(s, e), color="b")
```

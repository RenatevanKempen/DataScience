```python
import numpy as np
np.random.seed(123)

random_walk = [0]

for x in range(100):
    step = random_walk[-1]
    dice = np.random.randint(1,7)
    
    if dice <= 2:
        step = max(0, step -1)
    elif dice <= 5: #also noted by >3 and <6 
        step = step + 1
    else: 
        dice = step + np.random.randint(1,7)
    
    random_walk.append(step)

print(random_walk)
```

```python
import matplotlib.pyplot as plt
plt.plot(random_walk)
plt.show()
```

# Polygoon script voor GloBLD door BAIME

```python
# Voorbeeld met verzonnen data

import matplotlib.pyplot as plt
from matplotlib.collections import PolyCollection
import numpy as np

# some sample data
data = [(0,.5), (1,.7), (2,.3), (3,.6), (4,.2)]

fig = plt.figure()
ax = fig.add_subplot(111)

# draw the polygon
p = PolyCollection([data], facecolor=(1,0,0,.5), edgecolor='none')
ax.add_collection(p)

# draw the line
d = np.array(data)
ax.plot(*zip(*data), color='k', linewidth=2)
```

Super eenvoudig voorbeeld met als doel: laat zien dat je met Python een 3D polygoon kunt tekenen

```python
# Voorbeeld 3D polygoon
from mpl_toolkits import mplot3d
import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from itertools import product, combinations
```

```python
# Voorbeeld 3D polygoon (gekozen voor een eenvoudige kubus, maar allerlei vormen zijn mogelijk, zie hieronder)
fig = plt.figure()
ax = plt.axes(projection='3d')

r = [-1, 1]
for s, e in combinations(np.array(list(product(r, r, r))), 2):
    if np.sum(np.abs(s-e)) == r[1]-r[0]:
        ax.plot3D(*zip(s, e), color="b")
```

```python
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

x = np.linspace(-6, 6, 30)
y = np.linspace(-6, 6, 30)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
#Allerlei vormen zijn mogelijk! 
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap='binary')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```
